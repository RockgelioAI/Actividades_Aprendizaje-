### BoxPlot

```python
import seaborn as sns

sns.boxplot(x="drive-wheels", y="price", data=df)
```

### Scatter Plot

```python
import matplotlib.pyplot as plt

y=df['engine-size']
x=df['price']
plt.scatter(x,y)

plt.title("Scatterplot of engine size vs price")
plt.xlabel("Engine size")
plt.ylabel("Price")
```

### Groupby()

```python
df_test = df[['drive-wheels','body-style', 'price']]
df_grp = df_test.groupby(['drive-wheels','body-style'], as_index=False).mean()
df_grp
```

### Pivot - Method

```python
df_pivot = df_grp.pivot(index='drive-wheels', columns='body-style')
df_pivot
```

### Heatmap

```python
plt.pcolor(df_pivot, cmap='Paired')
plt.colorbar()
plt.show()
```

### ANOVA

```python
df_anova=df[["make","price"]]
grouped_anova=df_anova.groupby(["make"])
grouped_anova
```

```python
from scipy.stats import f_oneway

anova_results_1=f_oneway(grouped_anova.get_group("honda")["price"], grouped_anova.get_group("subaru")["price"])
anova_results_1
```

```python
anova_results_1=f_oneway(grouped_anova.get_group("honda")["price"], grouped_anova.get_group("jaguar")["price"])
anova_results_1
```
